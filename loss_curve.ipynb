```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Plotter():
    def __init__(self, data, legend):
        """
        data: list of lists of data to plot
        legend: list of strings to label each data set
        """
        self.data = data
        self.legend = legend
        self.t = len(self.data[0])

    def plot(self):
        for i in range(len(self.data)):
            plt.plot(self.data[i], label=self.legend[i])
        plt.plot([0, self.t], [0, 1], 'r--')
        plt.legend()

        plt.show()
```

```python
# Example
a = [1, 2, 4, 3.5, 5, 1, 2, 4, 3.5, 5]
a = np.array(a)/max(a)
b = [5, 4, 3, 2, 1, 1, 2, 4, 3.5, 5]
b = np.array(b)/max(b)
c = [2, 3, 4, 5, 6, 7, 8, 9, 10, 2]
c = np.array(c)/max(c)
data = [a, b, c]
plotter = Plotter(data, ['a', 'b', 'c'])
plotter.plot()
'''
The `Plotter` class takes a list of data in its constructor and has a `plot` method that uses `matplotlib` to plot the data. 
The `plot` method is called on an instance of the `Plotter` class to plot the data.
'''
```
