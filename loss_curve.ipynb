```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Plotter():
    def __init__(self, data_1, data_2, batch_size):
        """
        data: dictionary of lists of floats, where each list is a line to plot
        """
        self.data_1 = data_1
        self.data_2 = data_2
        self.t = batch_size

    def plot_acc(self, title, filename, labels):
        plt.plot(self.data_1['train_acc'], label=f'train_{labels[0]}')
        plt.plot(self.data_1['val_acc'], label=f'val_{labels[0]}')
        plt.plot(self.data_2['train_acc'], label=f'train_{labels[1]}', linestyle='dashed')
        plt.plot(self.data_2['val_acc'], label=f'val_{labels[1]}', linestyle='dashed')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.xticks(np.arange(0, self.t+1, step=5))
        plt.yticks(np.arange(0, 110, step=10))
        plt.title(title)
        plt.legend()
        plt.savefig(filename)
        plt.show()

    def plot_acc_2(self, title, filename, label):
        plt.plot(self.data_1['train_acc'], label=f'train_{label}')
        plt.plot(self.data_1['val_acc'], label=f'val_{label}')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.xticks(np.arange(0, self.t+1, step=5))
        plt.yticks(np.arange(0, 110, step=10))
        plt.title(title)
        plt.legend()
        plt.savefig(filename)
        plt.show()
    
    def plot_loss(self, title, filename, labels):
        plt.plot(self.data_1['train_loss'], label=f'train_{labels[0]}')
        plt.plot(self.data_1['val_loss'], label=f'val_{labels[0]}')
        plt.plot(self.data_2['train_loss'], label=f'train_{labels[1]}', linestyle='dashed')
        plt.plot(self.data_2['val_loss'], label=f'val__{labels[1]}', linestyle='dashed')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.xticks(np.arange(0, self.t+1, step=5))
        plt.title(title)
        plt.legend()
        plt.savefig(filename)
        plt.show()

    def plot_loss_2(self, title, filename, label):
        plt.plot(self.data_1['train_loss'], label=f'train_{label}')
        plt.plot(self.data_1['val_loss'], label=f'val_{label}')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.xticks(np.arange(0, self.t+1, step=5))
        plt.title(title)
        plt.legend()
        plt.savefig(filename)
        plt.show()
```
